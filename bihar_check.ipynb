```python
stat = {}

with open('list2.txt') as f:
    for l in f:
        fn = l.split('/')[-1]
        m = re.match("FinalRoll_ACNo_(\d+)PartNo_(\d+)\.pdf", fn)
        if m:
            ac = int(m.group(1))
            part = int(m.group(2))
            if ac in stat:
                stat[ac].append(part)
            else:
                stat[ac] = [part]
```
