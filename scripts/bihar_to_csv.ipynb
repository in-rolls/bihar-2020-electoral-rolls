```python
# gsutil -u in-electoral-rolls ls -al gs://in-electoral-rolls-2020/bihar > list3.txt
# FIXME: ignore last line, it's summary
df = pd.read_csv('../metadata_and_log_files/list3.txt', header=None, delimiter='\s+', nrows=72723, engine='python')
df.columns = ['filesize', 'timestamp', 'pdf_path', 'meta']
df
```

### Check smallest file size

```python
df.sort_values(['filesize'])
```

```python
def get_ac_part_no(r):
    for m in re.finditer('FinalRoll_ACNo_(\d+)PartNo_(\d+)\.pdf', r.pdf_path):
        ac = int(m.group(1))
        part = int(m.group(2))
        return ac, part
    print(r.pdf_path)
    return None, None

df[['ac_no', 'part_no']] = df.apply(lambda r: get_ac_part_no(r), axis=1, result_type='expand')
```

```python
df
```

```python
df = df.sort_values(['ac_no', 'part_no']).reset_index()
df
```

```python
mdf = pd.read_csv('../metadata_and_log_files/metadata.csv')
mdf = mdf.join(df)
odf = mdf[['ac_no', 'part_no', 'assembly_segment', 'part_number', 'pdf_path', 'filesize', 'timestamp']]
odf
```

```python
odf.to_csv('../metadata_and_log_files/bihar.csv', index=False)
```
