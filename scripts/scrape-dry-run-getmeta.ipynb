```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.select import Select
from IPython.core.display import Image, display
```

```python
options = Options()

options.add_experimental_option("prefs", {
      "download.default_directory": r"/opt/data/bihar-all",
      "download.prompt_for_download": False,
      "download.directory_upgrade": True,
      "safebrowsing.enabled": True
    })
 
options.add_argument('headless')
options.add_argument('--disable-infobars')
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--no-sandbox')
options.add_argument('--remote-debugging-port=9222')

driver = webdriver.Chrome('/opt/tools/webdriver/chromedriver', options=options)

driver.get('http://ele.bihar.gov.in/pdfsearch/')

# get the image source
```

```python
n = 0
sel = Select(driver.find_element_by_xpath('//select[@id="DropDownList1"]'))
a = 0
data = []
while a < len(sel.options):
    ac = sel.options[a].text
    sel.select_by_index(a)
    sel2 = Select(driver.find_element_by_xpath('//select[@id="DropDownList2"]'))
    c = len(sel2.options)
    n += c
    for d in sel2.options:
        data.append({'assembly_segment': ac, 'part_number': d.text})
    sel = Select(driver.find_element_by_xpath('//select[@id="DropDownList1"]'))
    print(a, c, n)
    a += 1
    #break
```

```python
driver.close()
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data)
df
```
